```python
import matplotlib.pyplot as plt
import data
DF30 = data.load_asset("SOLUSDT", sampling="30m")
DF30 = data.add_returns(DF30)

DF1d = data.load_asset("SOLUSDT", sampling="1d")
DF1d = data.add_returns(DF1d)
```

```python
fig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes = axes.flatten()

# Plot each graph with a title
DF30.iloc[300:350]["Return"].plot(ax=axes[0], title="DF30 Return")
DF30["Return"].rolling(window=5).mean().iloc[300:350].plot(ax=axes[1], title="DF30 Rolling Mean (5)")
DF30["Return"].cumsum().iloc[300:350].plot(ax=axes[2], title="DF30 Cumulative Return")
DF30["Return"].rolling(window=8).mean().cumsum().iloc[300:350].plot(ax=axes[3], title="DF30 Rolling Mean Cumulative (8)")

DF1d.iloc[300:350]["Return"].plot(ax=axes[4], title="DF1d Return")
DF1d["Return"].rolling(window=30).mean().iloc[300:350].plot(ax=axes[5], title="DF1d Rolling Mean (30)")
DF1d["Return"].cumsum().iloc[300:350].plot(ax=axes[6], title="DF1d Cumulative Return")
DF1d["Return"].rolling(window=4, center=False).mean().cumsum().iloc[300:360].plot(ax=axes[7], title="DF1d Rolling Mean Cumulative (4)")

# Adjust layout
plt.tight_layout()
plt.show()
```

As we can see, the rolling mean returns is very uneven, no matter the timescale... However, the cumulative sum of the rolling mean is very smooth and seems to have more trends...

Does this mean that it's easier to predict for the machine learning model? At the very least, if it predicts based on the rolling mean cumulative sum, it will be sort of aware of trends; will the cumulative sum continue to increase or decrease? (This is hopefully equivalent to asking will the returns be positive or negative)

HOWEVER; do notice that certain peaks are diminished with the rolling mean cumulative...

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import data as DATADATA

from keras.models import Sequential
from keras.layers import Dense, LSTM

# Scale the data
from sklearn.preprocessing import MinMaxScaler


for ticker in ['SOLUSDT', 'BTCUSDT', 'ETHUSDT', 'XRPUSDT', 'DOGEUSDT']:
    big_df = DATADATA.load_asset(ticker, sampling='1d')
    big_df['Returns'] = big_df['Close'].pct_change().rolling(window=4, center=False).mean().cumsum()
    big_df.dropna(inplace=True)
    scaler = MinMaxScaler(feature_range=(0,1))
    scaler.fit(big_df.filter(['Returns']))


    df = DATADATA.load_asset(ticker, sampling='1d')
    df = DATADATA.subset(df, start=pd.Timestamp('2022-01-01'), end=pd.Timestamp('2024-03-01'))


    # Create a new dataframe with only the 'Close column
    df['Returns'] = df['Close'].pct_change().rolling(window=4, center=False).mean().cumsum()
    df.dropna(inplace=True)
    data = df.filter(['Returns'])
    dataset = data.values

    # Convert the dataframe to a numpy array
    scaled_data = scaler.transform(dataset)

    scaled_data

    # Create the training data set
    # Create the scaled training data set
    train_data = scaled_data
    # Split the data into x_train and y_train data sets
    x_train = []
    y_train = []

    for i in range(60, len(train_data)):
        x_train.append(train_data[i-60:i, 0])
        y_train.append(train_data[i, 0])
        if i<= 61:
            print(x_train)
            print(y_train)
            print()

    # Convert the x_train and y_train to numpy arrays
    x_train, y_train = np.array(x_train), np.array(y_train)

    # Reshape the data
    x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
    # x_train.shape

    model = Sequential()
    model.add(LSTM(128, return_sequences=True, input_shape= (x_train.shape[1], 1)))
    model.add(LSTM(64, return_sequences=False))
    model.add(Dense(25))
    model.add(Dense(1))

    # Compile the model
    model.compile(optimizer='adam', loss='mean_squared_error')

    # Train the model
    model.fit(x_train, y_train, batch_size=8, epochs=16)

    # Get the models predicted price values
    predictions = model.predict(x_train)
    predictions = scaler.inverse_transform(predictions)

    # Get the root mean squared error (RMSE)
    rmse = np.sqrt(np.mean(((predictions - y_train) ** 2)))
    rmse


    valid = data[60:]
    valid['Predictions'] = predictions
    # display(valid.head(10))

    display(valid.plot())

    # Save the model
    model.save(f'models-KaggleSMA/lstm_model_rollmedicsum_{ticker}.keras')
```

Above, we can see that the model has fit the train data rather closely. How does it look on the test data?

```python
from keras.models import load_model

for ticker in ['SOLUSDT', 'BTCUSDT', 'ETHUSDT', 'XRPUSDT', 'DOGEUSDT']:
    big_df = DATADATA.load_asset(ticker, sampling='1d')
    big_df['Returns'] = big_df['Close'].pct_change().rolling(window=4, center=False).mean().cumsum()
    big_df.dropna(inplace=True)
    scaler = MinMaxScaler(feature_range=(0,1))
    scaler.fit(big_df.filter(['Returns']))


    df = DATADATA.load_asset(ticker, sampling='1d')
    df = DATADATA.subset(df, start=pd.Timestamp('2024-03-01'), end=pd.Timestamp('2026-01-01'))


    # Create a new dataframe with only the 'Close column
    df['Returns'] = df['Close'].pct_change().rolling(window=4, center=False).mean().cumsum()
    df.dropna(inplace=True)
    data = df.filter(['Returns'])
    dataset = data.values

    # Convert the dataframe to a numpy array
    scaled_data = scaler.transform(dataset)
    # Load the model
    model = load_model(f'models-KaggleSMA/lstm_model_rollmedicsum_{ticker}.keras')

    # Prepare the test data
    x_test = []
    y_test = dataset[60:]  # The actual returns
    for i in range(60, len(dataset)):
        x_test.append(scaled_data[i-60:i, 0])

    # Convert to numpy array and reshape
    x_test = np.array(x_test)
    x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))

    # Make predictions
    predictions = model.predict(x_test)
    predictions = scaler.inverse_transform(predictions)

    # Plot the predictions vs the real returns
    plt.figure(figsize=(10, 6))
    plt.plot(y_test, label="Real Returns")
    plt.plot(predictions, label="Predicted Returns")
    plt.legend()
    plt.title(f"Predictions vs Real Returns for {ticker}")
    plt.xlabel("Time")
    plt.ylabel("Returns")
    plt.show()
```
