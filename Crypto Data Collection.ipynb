```python
import pandas as pd
def response_to_dataframe(response, name):
    """Convert API response to Pandas DataFrame."""
    data = response.json()  # Convert response to JSON

    # Extract prices
    df = pd.DataFrame(data["prices"], columns=["timestamp", name])

    df["timestamp"] = df["timestamp"] - 86_400_000 

    # Convert timestamp from milliseconds to datetime
    df["timestamp"] = pd.to_datetime(df["timestamp"], unit="ms")

    return df
```

```python
def get_timeseries_data(name):
    url = "https://api.coingecko.com/api/v3/coins/" + name + "/market_chart/range?vs_currency=usd&from=1730433600&to=1739768400"
    headers = {
    "accept": "application/json",
    "x-cg-demo-api-key": "CG-ET5icj5F28nZUjNMKf4r5QJD"
}
    response = requests.get(url, headers=headers)
    df = response_to_dataframe(response, name)

    return df
    
get_timeseries_data("bitcoin")
```
